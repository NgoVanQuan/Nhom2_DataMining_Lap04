# 01 — Preprocessing & EDA (Beijing Multi-Site Air Quality)
Mục tiêu: tải dữ liệu, làm sạch, tạo nhãn phân lớp (AQI class theo PM2.5 24h mean), tạo đặc trưng thời gian + lag, và lưu `data/processed/cleaned.parquet`.

**Lưu ý:** nếu `USE_UCIMLREPO=True` thì notebook cần internet để tải dataset từ UCI.

```python
USE_UCIMLREPO = False
RAW_ZIP_PATH = "../data/raw/PRSA2017_Data_20130301-20170228.zip"

OUTPUT_CLEANED_PATH = '../data/processed/cleaned.parquet'
LAG_HOURS=[1, 3, 24]
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
import sys


project_root = Path('..').resolve()
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))
# --------------------------------

from src.classification_library import (
    load_beijing_air_quality,
    clean_air_quality_df,
    add_pm25_24h_and_label,
    add_time_features,
    add_lag_features,
)

PROJECT_ROOT = project_root # Tận dụng biến đã khai báo ở trên
OUT_PATH = (PROJECT_ROOT / OUTPUT_CLEANED_PATH).resolve()
OUT_PATH.parent.mkdir(parents=True, exist_ok=True)
```

```python
df_raw = load_beijing_air_quality(use_ucimlrepo=USE_UCIMLREPO, raw_zip_path=RAW_ZIP_PATH)
print('raw shape:', df_raw.shape)
df_raw.head()
```

```python
df = clean_air_quality_df(df_raw)
df = add_pm25_24h_and_label(df)
df = add_time_features(df)
df = add_lag_features(df, lag_hours=LAG_HOURS)
print('cleaned shape:', df.shape)
df[['datetime','station','PM2.5','pm25_24h','aqi_class']].head(10)
```

```python
# EDA nhanh: missingness và phân bố lớp
missing_rate = df.isna().mean().sort_values(ascending=False)
missing_rate.head(20)
```

```python
class_dist = df['aqi_class'].value_counts(dropna=False)
class_dist
```

```python
import matplotlib.pyplot as plt

class_dist.drop(index=[x for x in class_dist.index if pd.isna(x)], errors='ignore').plot(kind='bar')
plt.title('AQI class distribution (PM2.5 24h mean)')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Chọn 1 trạm để vẽ cho dễ nhìn (ví dụ: Aotizhongxin)
station_name = 'Aotizhongxin'
df_station = df[df['station'] == station_name].copy()
df_station = df_station.set_index('datetime').sort_index()

# 1. Vẽ toàn bộ giai đoạn
plt.figure(figsize=(15, 5))
plt.plot(df_station['PM2.5'], label='PM2.5 Raw', color='blue', alpha=0.5)
plt.title(f'Chuỗi PM2.5 toàn giai đoạn (2013-2017) - Trạm {station_name}')
plt.ylabel('PM2.5')
plt.legend()
plt.show()

# 2. Vẽ phóng to (Zoom) vào tháng 1/2016 (Mùa đông thường ô nhiễm cao)
subset = df_station.loc['2016-01-01':'2016-01-31']
plt.figure(figsize=(15, 5))
plt.plot(subset['PM2.5'], marker='o', linestyle='-', color='red')
plt.title(f'Chi tiết biến động PM2.5 tháng 01/2016 - Trạm {station_name}')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['PM2.5'])
plt.title('Phân bố giá trị PM2.5 (Boxplot)')
plt.show()
# Nhận xét: Các điểm đen phía bên phải là các giá trị ngoại lai (đỉnh ô nhiễm cực cao)
```

```python
from statsmodels.graphics.tsaplots import plot_acf

# Vẽ biểu đồ tự tương quan với 50 giờ trễ đầu tiên
plt.figure(figsize=(12, 6))
plot_acf(df_station['PM2.5'].dropna(), lags=50, ax=plt.gca())
plt.title('Autocorrelation Function (ACF) - Kiểm tra chu kỳ')
plt.xlabel('Lag (Giờ)')
plt.show()

# Insight: Nếu bạn thấy cột ở Lag 24 cao vọt lên, nghĩa là có chu kỳ ngày.
```

```python
from statsmodels.tsa.stattools import adfuller

# Chạy kiểm định ADF trên chuỗi PM2.5 (bỏ qua giá trị NaN)
result = adfuller(df_station['PM2.5'].dropna())

print('ADF Statistic:', result[0])
print('p-value:', result[1])

if result[1] < 0.05:
    print("=> Kết luận: Chuỗi có xu hướng DỪNG (Stationary).")
else:
    print("=> Kết luận: Chuỗi KHÔNG DỪNG (Non-Stationary).")
```

```python
df.to_parquet(OUT_PATH, index=False)
print('Saved:', OUT_PATH)
```
